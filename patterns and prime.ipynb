## check prime

```python
#program just for checking prime
n=int(input())
d=2
flag=False

while d<n:
    if(n % d ==0):
        flag==True
    d+=1
if flag==True:
    print("not prime")
else:
    print("prime")
    
```

```python
# printing all the primes till n
n=int(input())
k=2

while k<=n:
    d=2
    flag=False
    while d < k:
        if(k % d == 0):
            flag=True
        d=d+1
        
    if not(flag):
        print(k)
    k=k+1
```
