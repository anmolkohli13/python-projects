```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv("C:/Users/10times-61/Downloads/Kmeans_E3/E3/kmeans/student_evaluation.csv")

data.head()
```

```python
data.shape
```

```python
pd.isnull(data).sum()
```

```python
data.describe()
```

```python
kmeans=KMeans(n_clusters=2)
```

```python
kmeans.fit(data)
```

```python
pred=kmeans.predict(data)
```

```python
pred
```

```python
pd.Series(pred).value_counts()
```

```python
kmeans.inertia_
```

```python
kmeans.score(data)
```

```python
SSE=[]
```

```python
for cluster in range (1,20):
    kmeans=KMeans(n_jobs=-1,n_clusters=cluster)
    kmeans.fit(data)
    SSE.append(kmeans.inertia_)
```

```python
frame=pd.DataFrame({'Cluster':range(1,20),'SSE':SSE})
```

```python
frame.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'],frame['SSE'],marker="o")
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

data_scaled=scaler.fit_transform(data)
```

```python
pd.DataFrame(data_scaled).describe()
```

```python
SSE_scaled=[]
```

```python
for cluster in range (1,20):
    kmeans=KMeans(n_jobs=-1,n_clusters=cluster)
    kmeans.fit(data_scaled)
    SSE_scaled.append(kmeans.inertia_)
```

```python
frame_scaled=pd.DataFrame({'Cluster':range(1,20),'SSE':SSE_scaled})
plt.plot(frame_scaled['Cluster'],frame_scaled['SSE'],marker="o")
plt.xlabel("Clusters")
plt.ylabel("SSE")
```

```python
kmeans=KMeans(n_jobs=-1,n_clusters = 4)
kmeans.fit(data_scaled)
pred=kmeans.predict(data_scaled)
```
